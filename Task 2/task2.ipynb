```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv("fifa_eda_stats.csv")
```

```python
print(data.head())
```

```python
unwanted_columns = ['Loaned From','Joined','Contract Valid Until','Release Clause',
    'Preferred Foot',
    'Work Rate',
    'Body Type',
    'Jersey Number',
    'Height',
    'Weight',
    'International Reputation'
]
```

```python
data_cleaned = data.drop(columns=unwanted_columns)
```

```python
print(data_cleaned.head())
```

```python
#Check for missing values
print(data_cleaned.isna().sum())
```

```python
data_cleaned['Club'].fillna('No Club',inplace=True)
print(data_cleaned)
```

```python
fill_value=0
columns_to_fill = ['Weak Foot','Skill Moves', 'Position', 'Crossing', 'Finishing', 'HeadingAccuracy',     
'ShortPassing',        
'Volleys',             
'Dribbling', 'Curve',               
'FKAccuracy',         
'LongPassing',        
'BallControl',       
'Acceleration',     
'SprintSpeed',      
'Agility',             
'Reactions',           
'Balance',         
'ShotPower',           
'Jumping',         
'Stamina',             
'Strength',           
'LongShots',          
'Aggression',         
'Interceptions',       
'Positioning',     
'Vision',        
'Penalties',           
'Composure',           
'Marking',             
'StandingTackle',      
'SlidingTackle',       
'GKDiving',            
'GKHandling',          
'GKKicking',           
'GKPositioning',      
'GKReflexes' ]

data_cleaned[columns_to_fill] = data_cleaned[columns_to_fill].fillna(fill_value)
print(data_cleaned.head())
```

```python
print(data_cleaned.dtypes)
```

```python
#Data type conversion
data_cleaned['Value'] = data_cleaned['Value'].str.replace('€', '')
data_cleaned['Wage'] = data_cleaned['Wage'].str.replace('€', '')

data_cleaned['Value'] = data_cleaned['Value'].str.replace('K','e3').str.replace('M','e6').astype(float)
data_cleaned['Wage'] = data_cleaned['Wage'].str.replace('K','e3').str.replace('M','e6').astype(float)
```

```python
# Remove duplicate rows
data_cleaned.drop_duplicates(inplace=True)
```

```python
#normalising the data 
import sklearn
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
numerical_columns = ['Overall', 'Potential', 'Crossing', 'Finishing', 'ShortPassing', 'Dribbling', 'BallControl', 'Acceleration', 'SprintSpeed', 'Agility', 'Reactions', 'Balance', 'ShotPower', 'Jumping', 'Stamina', 'Strength', 'LongShots', 'Aggression', 'Interceptions', 'Positioning', 'Vision', 'Penalties', 'Composure', 'Marking', 'StandingTackle', 'SlidingTackle', 'GKDiving', 'GKHandling', 'GKKicking', 'GKPositioning', 'GKReflexes']
data_cleaned[numerical_columns] = scaler.fit_transform(data_cleaned[numerical_columns])
```

```python
#Save the cleaned and preprocessed data
data_cleaned.to_csv('cleaned_fifa_data.csv', index=False)

```

```python
#Age Categories
import matplotlib.pyplot as plt

df = pd.read_csv("cleaned_fifa_data.csv")

Ages = df['Age']

# Define age ranges and their corresponding labels
age_ranges = ["11-20", "21-30", "31-40", "41-50"]
age_counts = [0, 0, 0, 0]

# Count the ages in their respective ranges
for age in Ages:
    if age <= 20:
        age_counts[0] += 1
    elif age <= 30:
        age_counts[1] += 1
    elif age <= 40:
        age_counts[2] += 1
    else:
        age_counts[3] += 1

plt.figure(figsize=(8, 8))
plt.pie(age_counts,labels=age_ranges, startangle = 140, autopct='%1.1f%%')
plt.title("Age Distribution")
plt.show()
```
